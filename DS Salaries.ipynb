```python
def countZeroYesNo(target_data):
    count_US = 0
    count_NONUS = 0
    for i in range(len(target_data)):
        if(target_data['company_location'].tolist()[i] == 'US'):
            count_US += 1
        else:
            count_NONUS += 1
    print(count_US,count_NONUS)
    if(count_US > count_NONUS):
        return 'Its in US'
    else:
        return 'Its non-US'
predictor = countZeroYesNo(df)
predictor
```

```python
def countYesNo(target_data):
    count_yes = 0
    count_no = 0
    for i in range(len(target_data)):
        if(target_data['employee_residence'].tolist()[i] == 'US' and target_data['salary_currency'].tolist()[i] == 'USD'):
            count_yes += 1
        else:
            count_no += 1
    print(count_yes, count_no)
    if(count_yes > count_no):
        return 'Employee is American'
    else:
        return 'Employee is not American'
countYesNo(df)
```

```python
def findVariance(data, mean):
    variance = 0
    for i in data:
        variance += (i - mean)*(i - mean)
    return variance/(len(data)-1)
```

```python
var1 = findVariance(df[data_x]['salary'], mean1)
var2 = findVariance(df[data_x]['remote_ratio'], mean2)
var3 = findVariance(df[data_y]['salary'], mean3)
var4 = findVariance(df[data_y]['remote_ratio'], mean4)
print(var1, var2)
print(var3, var2)
```

```python
def getProbabilityDistribution(mean, variance, 
                               target_variable):
    p = (1.0/math.sqrt(2*math.pi*variance))* math.exp( 
    ((-1)*(target_variable-mean)*(target_variable-mean))
    /(2*variance)
    )
    return p
```

```python
p_salary_x = getProbabilityDistribution(mean1, var1, target_data['salary'][0])
p_remoteratio_x = getProbabilityDistribution(mean2, var2, target_data['remote_ratio'][0])
p_salary_y = getProbabilityDistribution(mean3, var3, target_data['salary'][0])
p_remoteratio_y = getProbabilityDistribution(mean4, var4, target_data['remote_ratio'][0])
```
